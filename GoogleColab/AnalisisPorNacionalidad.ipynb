```python
#Carga de Librerias
#Visualizacion de datos
import pandas as pd
#Trabajar con excel
import openpyxl
##Libreria para las grafica
import matplotlib.pyplot as plt

import plotly.graph_objs as go

import plotly.express as px

import numpy as np

import plotly.graph_objs as go
```

```python
#Carga del dataset y Visualizacion de datos
#Ruta de archivo
ruta_archivo = '/content/dataSet.xlsx'

#Validar si el archivo existe
#!ls '/content/dataSetPrueba.xlsx',nrows=10000
#Leer el archivo
datos = pd.read_excel(ruta_archivo)
datos.head(7)
```

```python
#Procesamiento de Datos
#Chile
suma_hombres_cl = datos['Chilena'].sum()
suma_mujeres_cl = datos['Chilena'].size - np.count_nonzero(datos['Chilena'])
total_Chilenos=suma_hombres_cl+suma_mujeres_cl
print (suma_hombres_cl)
print (suma_mujeres_cl)
print ("Chilenos", total_Chilenos)

#Venezolanos
suma_hombres_ve = datos['Venezolana'].sum()
suma_mujeres_ve = datos['Venezolana'].size - np.count_nonzero(datos['Venezolana'])
total_venezolanos=suma_hombres_ve+suma_mujeres_ve
print (suma_hombres_ve)
print (suma_mujeres_ve)
print ("Venezuela", total_venezolanos)

#Bolivia
suma_hombres_bo = datos['Boliviana'].sum()
suma_mujeres_bo = datos['Boliviana'].size - np.count_nonzero(datos['Boliviana'])
total_bolivia=suma_hombres_bo+suma_mujeres_bo
print (suma_hombres_bo)
print (suma_mujeres_bo)
print ("Bolivia", total_bolivia)

#Colombiana
suma_hombres_co = datos['Colombiana'].sum()
suma_mujeres_co = datos['Colombiana'].size - np.count_nonzero(datos['Colombiana'])
total_colombia=suma_hombres_co+suma_mujeres_co
print (suma_hombres_co)
print (suma_mujeres_co)
print ("Colombia", total_colombia)

#Peruana
suma_hombres_pe = datos['Peruana'].sum()
suma_mujeres_pe = datos['Peruana'].size - np.count_nonzero(datos['Peruana'])
total_peru=suma_hombres_pe+suma_mujeres_pe
print (suma_hombres_pe)
print (suma_mujeres_pe)
print ("Peru", total_peru)

#Haitiana
suma_hombres_hai = datos['Haitiana'].sum()
suma_mujeres_hai = datos['Haitiana'].size - np.count_nonzero(datos['Haitiana'])
total_haiti=suma_hombres_hai+suma_mujeres_hai
print (suma_hombres_hai)
print (suma_mujeres_hai)
print ("Haiti", total_haiti)

#Ecuatoriana
suma_hombres_ec = datos['Ecuatoriana'].sum()
suma_mujeres_ec = datos['Ecuatoriana'].size - np.count_nonzero(datos['Ecuatoriana'])
total_ecuador=suma_hombres_ec+suma_mujeres_ec
print (suma_hombres_ec)
print (suma_mujeres_ec)
print ("Ecuador", total_ecuador)

#Argentina
suma_hombres_ar = datos['Argentina'].sum()
suma_mujeres_ar = datos['Argentina'].size - np.count_nonzero(datos['Argentina'])
total_argentina=suma_hombres_ar+suma_mujeres_ar
print (suma_hombres_ar)
print (suma_mujeres_ar)
print ("Argentina", total_argentina)


#Total Todo menos Chilenos
total_ex = total_venezolanos+total_bolivia+total_colombia+total_peru+total_haiti+total_ecuador+total_argentina




```

```python
datosnuevos = {
    'Nacionalidad': ['Chile','Venezuela','Boliviana','Colombiana','Peruana','Haitiana','Ecuatoriana','Argentina'],
    'Hombre': [suma_hombres_cl, suma_hombres_ve, suma_hombres_bo, suma_hombres_co, suma_hombres_pe, suma_hombres_hai, suma_hombres_ec, suma_hombres_ar],
    'Mujer': [suma_mujeres_cl, suma_mujeres_ve, suma_mujeres_bo, suma_mujeres_co, suma_mujeres_pe, suma_mujeres_hai, suma_mujeres_ec, suma_mujeres_ar]
    }

newDt = pd.DataFrame(datosnuevos)

newDt.head(7)
```

```python
newDt.head(5)
```

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Crear los subplots para los dos anillos
plt.subplot(1, 2, 1)
plt.pie(newDt['Hombre'], labels=newDt['Nacionalidad'], autopct='%6.1f%%', startangle=10)
plt.title('Total Hombres')
# Agregar un círculo blanco en el centro para hacerlo un diagrama de anillo
centro_circulo = plt.Circle((0, 0), 0.7, color='white')
fig = plt.gcf()
fig.gca().add_artist(centro_circulo)

plt.subplot(1, 2, 2)
plt.pie(newDt['Mujer'], labels=newDt['Nacionalidad'], autopct='%1.1f%%', startangle=90)
plt.title('Total Mujeres')

# Agregar un círculo blanco en el centro para hacerlo un diagrama de anillo
centro_circulo = plt.Circle((0, 0), 0.7, color='white')
fig = plt.gcf()
fig.gca().add_artist(centro_circulo)

# Ajustar la disposición de los subplots
plt.tight_layout()

# Mostrar el diagrama de doble anillo
plt.show()
```

```python
# Datos para el diagrama de pai
etiquetas = ['Chilenos', 'Extranjeros']  # Etiquetas para cada sección
valores = [total_Chilenos, total_ex]  # Valores correspondientes a cada sección
colores = [ 'green', 'orange']  # Colores para cada sección

# Crear el diagrama de pai
plt.figure(figsize=(6, 6))
plt.pie(valores, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)

# Ajustar el aspecto
plt.axis('equal')

# Mostrar el diagrama de pai
plt.title('Diagrama de Pay')
plt.show()
```

```python
# Datos para el diagrama de barras con la Nacionalidad
nacionalidad = ['Chilena','Venezolana','Boliviana','Colombiana','Peruana','Haitiana','Ecuatoriana','Argentina ']  # Categorías en el eje x
valores = [total_Chilenos, total_venezolanos, total_bolivia, total_colombia, total_peru, total_haiti, total_ecuador, total_argentina ]  # Valores correspondientes a cada categoría

# Crear el diagrama de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura (opcional)
plt.bar(nacionalidad, valores, color='skyblue')

# Añadir etiquetas y título
plt.xlabel('Nacionalidades')
plt.ylabel('')
plt.title('Diagrama de Barras con la Nacionalidad')

# Mostrar el diagrama de barras
plt.show()
```
